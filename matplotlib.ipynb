# Let's plot out first graph!
Here is a representation of how you make a line graph using matplotlib. The functions used are self explanatory and you can easily understand them

```python
x=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
y=[150,120,100,90,200,70,110]

plt.plot(x,y) # to plot your data

plt.title("Chocolate Sales in a week")  # adds title to your figure

plt.xlabel("Days") # gives lable your x-axis

plt.ylabel("Sales per day") # gives lable your y-axis

plt.grid() # adds a gridline at the back

plt.show() # used to show your chart
```

Now, let's study these functions individually for more understanding along with some other functions!

## Plot
You can give multiple parameters to the **.plt** function like:\
* linestyle
* color
* marker


```python
x=[1,2,3]
y=[12,1,23]
plt.plot(x,y,color="red",marker="o",linestyle="--",linewidth=1,label="Sample graph")
plt.show()
```

## Title and x-y Labels
Self explainatory, Title adds a title to your graph that tells what your graph is about in a few words and x and y labels tell what data in represented on x axis and y axis respectively.

```python
x=[1,2,3]
y=[12,1,23]
plt.plot(x,y,color="red",marker="o",linestyle="--",linewidth=1)
plt.title("Title")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

## Legend and Grid
It is like a label box that tells what represents what\
You can give parameters indicating the location and font size 
> Grid can be added to the back of your graph for better readability and you can also specify the line style and color as of the plot.

```python
Months=[1,2,3,4]
Sales=[2000,1000,1500,3000]
plt.plot(Months,Sales,color="red",marker="o",linestyle="--",linewidth=1)
plt.title("Sales data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(fontsize=11,loc="upper left")
plt.grid(color="gray",linestyle=":")
plt.show()
```

## Axis-ticks
It is used for giving meaningful labeling, as in the above graph months are represented in numbers but we can also change it for better understanding

```python
Months=[1,2,3,4]
Sales=[2000,1000,1500,3000]
plt.plot(Months,Sales,color="red",marker="o",linestyle="--",linewidth=1)
plt.title("Sales data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(fontsize=11,loc="upper left")
plt.grid(color="gray",linestyle=":")


plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

Similarly, You can do the same for y-axis using **yticks**.

## Axis limits
xlim and ylim are used to focus on a specific part of the graph rather than showing whole graph.

```python
Months=[1,2,3,4]
Sales=[2000,1000,1500,3000]
plt.plot(Months,Sales,color="red",marker="o",linestyle="--",linewidth=1)
plt.title("Sales data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(fontsize=11,loc="upper left")
plt.grid(color="gray",linestyle=":")
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.xlim(1,4)


plt.ylim(100,2000)
plt.show()
```

## Bar chart
A bar chart is used when you want to compare quantities across different categories. Itâ€™s best for visualizing discrete data

```python
#syntax
# plt.bar(x,y,color="color",width=value,label="labelname")

products=["Apple","Orange","Mango","Watermelon"]
sales=[1200,560,2000,1800]
plt.bar(products,sales,color="yellow",label="Sales Data")
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

## Pie chart
Pie charts are used when you want to show how a whole is divided into parts, especially to highlight the proportions or percentages of different categories within a single dataset

```python
#Syntax
#plt.pie(values,labels=label_list,color=[color list],autopct='%1.1f%%')
```

> In   **autopct='%1.1f%%'**   %1,1 means one digit before the decimal point and one after, %% will show a '%' sybmol after the numerical value

```python
region=['north','south','west','east']
sales=[3000,2000,1500,700]
plt.pie(sales,labels=region,autopct='%1.1f%%',colors=['gold','lightblue','cyan','coral'])
plt.title("Revenue by region")
plt.show()
```

## Histogram
A histogram is used to show the distribution of numerical data by grouping it into ranges

```python
#syntax
#plt.hist(data,bins=number_of_bins,color='colorname',edgecolor='colorname')
```

```python
marks=[50,45,67,57,87,88,83,76,90,96,64,49,55,79,83,72,89]
plt.hist(marks,bins=5,color='lightblue',edgecolor='black')
plt.title("Student marks distribution")
plt.xlabel("Score range")
plt.ylabel("Students")
plt.show()
```

> Height og bar = number of students that fall in that range

## Scatter plot
A scatter plot is used when you want to visualize the relationship or correlation between two numerical variables.It is used to identify trends, patterns, or outliers

```python
#syntax
#plt.scatter(x,y,color='colorname',marker='symbol',label='labelname')


hours_studied=[1,2,3,4,5,6,7]
score=[30,40,50,60,70,80,90]
plt.scatter(hours_studied,score,marker='o',color='blue',label='Student data')
plt.xlabel("Hours studied")
plt.ylabel("Score")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

Similarly,we can plot a scatter chart for showing relation between two variables like this:

```python
hours_studied=[1,2,3,4,5]
marksg1=[45,50,55,57,60]
marksg2=[50,49,39,70,67]
plt.scatter(hours_studied,marksg1,marker='o',color='blue',label='Group 1')
plt.scatter(hours_studied,marksg2,marker='o',color='red',label='Group 2')
plt.xlabel("Hours studied")
plt.ylabel("Score")
plt.legend(loc='upper left')
# plt.grid(True)
plt.show()
```

## Subplots and Layout Adjustments
A subplot is a smaller plot within a larger figure that allows you to display multiple graphs (plots) side by side or in a grid layout
There are two ways to make subplots. First is quick and easier:

```python
#syntax
#plt.subplot(nrows,ncols,index)
x=[1,2,3,4,5]
y=[20,17,10,29,40]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar chart")
plt.show()
```

The other method is preffered more in industry 

## Object Oriented Way

this is a way that gives you more command while creating your plots and graphs

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.6]) # left,bottom,width,height
axis1.plot(x,y)
```

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.6]) # left,bottom,width,height
axis1.plot(x,y)
axis2=fig.add_axes([0.5,0.5,0.2,0.3]) # left,bottom,width,height
axis2.plot(x,y,color='red')
```

Thus, Object oriented way gives you liberty to manipulate your canvas according to your will and move plots

## Saving Figures

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.6]) # left,bottom,width,height
axis1.plot(x,y)
fig.savefig("myfigure.png")  # saves the figure in the current working directory
```

## Distribution plot
Distribution plot works on quantitative data, we'll discuss some of these plots in this notebook

We will be discussing:
* hisplot/distplot
* jointplot
* pairplot
* rugplot

```python
import seaborn as sns
```

Seaborn allows you to import dataset directly from its dataset library available like this:

```python
df=sns.load_dataset('tips')
df
```

as we will be discussing distribution plots ,we need only quantitative data. In this dataset, quantitative columns are total_bill and tip only

```python
sns.displot(df['total_bill'])
#displot and histogram are same, means you the code below will give you same output
sns.histplot(df['total_bill'])
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],kde=True)  
plt.subplot(1,2,2)
sns.histplot(df['tip'],kde=True,color='red')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```

```python
sns.rugplot(df["total_bill"])
```
